```python
import csv
import numpy as np
import pandas as pd
```

# Faculty to Teaching Areas

```python
df_fullFaculty = pd.read_csv('Data_FullFaculty.csv')
df_fullFaculty
```

```python
list_ids = []
list_titles = []
list_areas = []

for i in range(len(df_fullFaculty)):
    temp_id = df_fullFaculty['ID'][i]
    temp_title = df_fullFaculty['Title'][i]
    
    list_rowAreas = df_fullFaculty['Teaching Areas'][i].split('|')
    num_rowAreas = len(list_rowAreas)
    
    for j in range(num_rowAreas):
        list_ids.append(temp_id)
        list_titles.append(temp_title)
        list_areas.append(list_rowAreas[j])
    
```

```python
df_facultyAreas = pd.DataFrame(list(zip(list_ids, list_titles, list_areas)), columns =['ID', 'Title', 'Teaching Areas']) 
df_facultyAreas
```

```python
df_facultyAreas.to_csv('Data_FacultyAreas.csv', index=False)
```
